## Problem-1(a) visualization - Customer Behaviour

```python
# creating a 2D plot of the customer data. 
_, ax = plt.subplots(figsize=(7, 5))
# select all the diff colours reds & blues
blues = customer_df[customer_df["c"] == 0]
reds = customer_df[customer_df["c"] == 1]
# pull into scatter plots

ax.scatter(x=blues["x1"], y=blues["x2"], c="blue", label='y=0 (Not Interested)', s=80, zorder=3)
ax.scatter(x=reds["x1"], y=reds["x2"], c="red", label='y=1 (Interested)', s=80, zorder=3)

## Plot it out & show:
ax.set_xlabel('x1: Avg Amount Spent per Purchase ($)')
ax.set_ylabel('x2: Frequency of Purchases per Month')
ax.set_title('Q1a: Customer Data Scatter Plot')
ax.legend()
ax.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()

```
